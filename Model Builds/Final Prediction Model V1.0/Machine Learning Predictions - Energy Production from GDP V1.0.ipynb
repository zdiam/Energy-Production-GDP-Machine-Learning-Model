```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import pickle
%matplotlib inline
```

```python
modelKN = pickle.load(open('KNeigh.pkl', 'rb'))
```

```python
dataP = pd.read_csv (r'dataP.csv')
```

```python
dataP = dataP[['LOCATION','TIME','Value']]
```

```python
dataP
```

```python
dataCanada = dataP.iloc[0:41]
XCanada = dataCanada.iloc[:,2].values.reshape(-1,1)
yCanada = modelKN.predict(XCanada)
```

```python
plt.scatter(XCanada,yCanada)
plt.title('Canada Prediction')
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```
