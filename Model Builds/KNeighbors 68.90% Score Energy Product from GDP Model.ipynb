```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataW = pd.read_csv (r'dataW.csv')
```

```python
dataW
```

```python
dataW.dtypes
```

```python
X = dataW.iloc[:,3].values.reshape(-1,1) #GDP will be X to use predict the Y Energy Production
```

```python
X
```

```python
X.shape
```

```python
Y = dataW.iloc[:,2].values.reshape(-1,1) 
```

```python
Y
```

```python
Y.shape
```

```python
plt.scatter(X,Y)
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=10)
```

```python
modelKN = neigh.fit(X,Y)
```

```python
modelKN.score(X,Y)
```

```python
yPredKN = modelKN.predict(X)
plt.scatter(X,Y)
plt.plot(X,yPredKN,color='red')
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```

```python
#Training model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=.3,random_state = 4444 )
```

```python

modelKN = neigh.fit(X_train,Y_train)
Y_test_pred = modelKN.predict(X_test)
plt.scatter(X_test,Y_test)
plt.plot(X_test,Y_test_pred,color='red')
plt.xlabel('GDP')
plt.ylabel('Energy Production Quad BTU')
```

```python
modelKN.score(X_test,Y_test)
```
